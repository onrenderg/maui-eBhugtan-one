```python
class Payments
{
     These names must match your JSON keys exactly
    public string BillID { get; set; }
    public string BillType { get; set; }
    public string DDOCode { get; set; }
    public decimal NetAmount { get; set; } // We can use decimal for money!
    public string PaidOn { get; set; }
    public string PayeeCode { get; set; }
    public string Remarks { get; set; }
    public string Treasury { get; set; }
}

public class PaymentsDatabase

{
    public static string  Dbname = "payments.db";

    private SQLiteConnection _conn;

    public PaymentsDatabase()
    {
        var dbPath = Path.Combine(Environment.CurrentDirectory, Dbname);
        _conn = new SQLiteConnection(dbPath);
        _conn.CreateTable<Payments>();
    }


    // Read all payments from the database
    public IEnumerable<Payments> GetPayments()
    {
        var list = _conn.Query<Payments>("SELECT * FROM Payments");
        return list.ToList();
    }

    
}

PaymentsDatabase database_Payments = new PaymentsDatabase();

Console.WriteLine(database_Payments.GetPayments().Count());
```
