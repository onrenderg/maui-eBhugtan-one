```python
import requests
import json

class HitServices:
    # Defining variables as class attributes
    encrypted_param = "7JsZWOtl65fM6VjfNVIfnHPQeRaRLQWTbodwGqmPGRLkeAwCes5VCFF9VqdltRATKe0udbpaDSnx0sONsGYmkPEWo4i3SKEh2wxdzQOgKcdSxOoAPByIL6famC4goGjnKmQIfzufhd7aT1forfd7zT8l2fEwOnom7nA8ZaeIQag="
    base_url = f"http://10.146.15.95/eHPOLTIS/eBhugtanService/eBhugtan.svc/GetBillDetails?param={encrypted_param}"

    def get_hit(self):
        # Use self.base_url to access the variable above
        response = requests.get(self.base_url)
        
        if response.status_code == 200:
            # Use .json() if the API returns JSON data
            return response.json() 
        else:
            raise Exception(f"Failed: {response.status_code} - {response.text}")

# Initialize and run
hist_services = HitServices()
hit_data = hist_services.get_hit()

# THIS IS THE PART THAT PRINTS THE DATA
print(json.dumps(hit_data, indent=4))
```
